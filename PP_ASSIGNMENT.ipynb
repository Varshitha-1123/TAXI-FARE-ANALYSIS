```python
# proj01.py
# Author: [Your Name]
# CSE 231, Spring 2017 - Project 1
# This program converts distance in rods to meters, feet, miles, furlongs,
# and calculates the time to walk that distance at 3.1 miles/hour.

# Constants
ROD_TO_METERS = 5.0292
METERS_TO_FEET = 1 / 0.3048
METERS_TO_MILES = 1 / 1609.34
RODS_TO_FURLONGS = 1 / 40
WALKING_SPEED_MPH = 3.1

# Input from user
rods_str = input("Input distance in rods: ")
rods = float(rods_str)

# Calculations
meters = rods * ROD_TO_METERS
feet = meters * METERS_TO_FEET
miles = meters * METERS_TO_MILES
furlongs = rods * RODS_TO_FURLONGS
hours = miles / WALKING_SPEED_MPH
minutes = hours * 60

# Output
print(f"\nDistance in rods: {rods}")
print(f"Distance in meters: {meters:.3f}")
print(f"Distance in feet: {feet:.3f}")
print(f"Distance in miles: {miles:.6f}")
print(f"Distance in furlongs: {furlongs:.3f}")
print(f"Time to walk {rods} rods: {minutes:.1f} minutes")
```

```python
# proj01.py
# Author: [Your Name]
# CSE 231, Fall 2013 - Project 1
# This program calculates the energy released by an earthquake given its Richter scale value.
# It displays the result in joules and equivalent tons of exploded TNT.

# Constants
TNT_ENERGY_JOULES = 4.184e9

# Function to calculate energy in joules
def calculate_energy(richter):
    return 10 ** (1.5 * richter + 4.8)

# Function to convert energy to TNT tons
def energy_to_tnt(energy):
    return energy / TNT_ENERGY_JOULES

# Predefined Richter values
richter_values = [1.0, 5.0, 9.1, 9.2, 9.5]

# Process and display results for predefined values
print("Richter  Energy (Joules)     Tons of TNT")
print("-------------------------------------------")
for r in richter_values:
    energy = calculate_energy(r)
    tnt = energy_to_tnt(energy)
    print(f"{r:<7} {energy:>13.3e} {tnt:>17.3e}")

# Prompt user for input
user_input = input("\nEnter a Richter scale value: ")
richter_value = float(user_input)

# Compute and display result for user input
user_energy = calculate_energy(richter_value)
user_tnt = energy_to_tnt(user_energy)
print(f"\nRichter value: {richter_value}")
print(f"Energy released: {user_energy:.3e} joules")
print(f"Equivalent to: {user_tnt:.3e} tons of TNT")
```

```python
import math

# Constants
speed_of_light = 299792458  # meters per second
mass_of_ship = 70000  # kilograms (at rest)

# Prompt the user for the percentage of the speed of light
percent_speed = float(input("Enter the percentage of the speed of light your ship will travel: "))

# Convert percentage to velocity
velocity = (percent_speed / 100) * speed_of_light

# Calculate the time dilation factor using the relativistic formula
time_dilation_factor = 1 / math.sqrt(1 - (velocity / speed_of_light) ** 2)

# Calculate the mass increase
increased_mass = mass_of_ship * time_dilation_factor

# Calculate the times for various trips (in years)
# Convert light years to time in years using the time dilation factor
def calculate_travel_time(distance_light_years):
    travel_time = distance_light_years / (velocity / speed_of_light)  # time in years
    astronaut_time = travel_time / time_dilation_factor  # time experienced by astronauts
    return astronaut_time

# Distances in light years
distances = {
    "Alpha Centauri": 4.3,
    "Barnard's Star": 6.0,
    "Betelgeuse": 309,
    "Andromeda Galaxy": 2000000
}

# Output the results
print(f"Mass of the ship: {mass_of_ship} kg")
print(f"Mass of the ship at {percent_speed}% of the speed of light: {increased_mass} kg")
print()

for star, distance in distances.items():
    astronaut_time = calculate_travel_time(distance)
    print(f"Time to {star} for astronauts: {astronaut_time:.2f} years")
```

```python
import math

def main():
    # Prompt the user for input
    garden_length = float(input("Enter the length of one side of the garden (in feet): "))
    plant_spacing = float(input("Enter the recommended spacing between plants (in inches): "))
    flowerbed_depth = float(input("Enter the depth of the flowerbeds (in feet): "))
    fill_depth = float(input("Enter the depth of the fill areas (in feet): "))

    # Convert spacing from inches to feet
    plant_spacing_feet = plant_spacing / 12

    # Calculate the radius of the circle flowerbed (half of the garden's side length)
    radius = garden_length / 2

    # Calculate the area of the central circular flowerbed
    circle_area = math.pi * radius * radius

    # Calculate the area of one outer triangular flowerbed
    triangle_base = garden_length  # The base of the triangle is the garden's side length
    triangle_height = garden_length / 2  # The height of the triangle is half the garden's side length
    triangle_area = 0.5 * triangle_base * triangle_height

    # Calculate the number of plants for each flowerbed type
    area_per_plant = plant_spacing_feet ** 2
    num_plants_circle = int(circle_area / area_per_plant)
    num_plants_triangle = int(triangle_area / area_per_plant)

    # Total number of plants (4 triangles + 1 circle)
    total_plants = 4 * num_plants_triangle + num_plants_circle

    # Calculate the volume of soil (in cubic yards) for each flowerbed type
    # Volume = Area * Depth, then convert from cubic feet to cubic yards (1 cubic yard = 27 cubic feet)
    soil_volume_circle = (circle_area * flowerbed_depth) / 27
    soil_volume_triangle = (triangle_area * flowerbed_depth) / 27
    total_soil_volume = 4 * soil_volume_triangle + soil_volume_circle

    # Calculate the volume of fill material (in cubic yards)
    # Total area = garden_length^2 (since the garden is square), subtract areas of flowerbeds
    total_area = garden_length ** 2
    total_flowerbed_area = 4 * triangle_area + circle_area
    fill_area = total_area - total_flowerbed_area
    fill_volume = (fill_area * fill_depth) / 27

    # Output the results
    print("\nGarden Material Estimates:")
    print(f"Number of plants in the circular flowerbed: {num_plants_circle}")
    print(f"Number of plants in each triangular flowerbed: {num_plants_triangle}")
    print(f"Total number of plants for the garden: {total_plants}")

    print(f"Cubic yards of soil for the circular flowerbed: {soil_volume_circle:.1f}")
    print(f"Cubic yards of soil for each triangular flowerbed: {soil_volume_triangle:.1f}")
    print(f"Total cubic yards of soil for the garden: {total_soil_volume:.1f}")

    print(f"Total cubic yards of fill material for the garden: {fill_volume:.1f}")

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import math

def calculate_angle(x1, y1, x2, y2):
    """Calculates the angle between two lines defined by points (0, 0), (x1, y1) and (x1, y1), (x2, y2)."""

    # Calculate the slope of the first line (from (0,0) to (x1, y1))
    m1 = y1 / x1 if x1 != 0 else float('inf')

    # Calculate the slope of the second line (from (x1, y1) to (x2, y2))
    m2 = (y2 - y1) / (x2 - x1) if (x2 - x1) != 0 else float('inf')

    # Calculate the angle between the two lines using the formula
    # tan(theta) = |(m1 - m2) / (1 + m1 * m2)|
    tan_theta = abs((m1 - m2) / (1 + m1 * m2))

    # Calculate the angle in radians and then convert to degrees
    theta_radians = math.atan(tan_theta)
    theta_degrees = math.degrees(theta_radians)

    return theta_degrees

def main():
    # Print a description of the program
    print("Welcome to the Geometry and Matplotlib Plotting Program!")
    print("This program will plot two lines and calculate the acute angle between them.")

    # Prompt the user for the coordinates of the first point
    x1 = float(input("Enter the x coordinate of the first point: "))
    y1 = float(input("Enter the y coordinate of the first point: "))

    # Prompt the user for the coordinates of the second point
    x2 = float(input("Enter the x coordinate of the second point: "))
    y2 = float(input("Enter the y coordinate of the second point: "))

    # Calculate the angle between the two lines
    angle = calculate_angle(x1, y1, x2, y2)

    # Print the angle
    print(f"The acute angle between the two lines is {angle:.2f} degrees.")

    # Plot the lines using matplotlib
    plt.figure(figsize=(6, 6))  # Set figure size
    plt.plot([0, x1, x2], [0, y1, y2], marker='o', linestyle='-')  # Plot lines with markers

    # Annotate the points
    plt.text(x1, y1, f'({x1:.2f}, {y1:.2f})', ha='right', va='bottom')
    plt.text(x2, y2, f'({x2:.2f}, {y2:.2f})', ha='left', va='top')

    # Annotate the angle
    plt.text((x1 + x2) / 2, (y1 + y2) / 2, f'Angle: {angle:.2f}°', ha='center', va='center')

    # Set plot limits and labels
    plt.xlim(-10, 10)  # Adjust limits as needed
    plt.ylim(-10, 10)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Geometry and Matplotlib Plotting")
    plt.grid(True)

    # Show the plot
    plt.show()

# Call the main function to run the program
if __name__ == "__main__":
    main()
```

```python
# proj03.py - Whole number splitting into pieces in increasing order

# Function to check if a given string can be split into pieces of length 'n' and if they are in increasing order
def check_increasing_order(number, piece_length):
    # Split the number into pieces of length 'piece_length'
    pieces = []
    for i in range(0, len(number), piece_length):
        pieces.append(int(number[i:i + piece_length]))

    # Print the pieces
    print("Pieces:", ', '.join(map(str, pieces)))

    # Check if the pieces are in increasing order
    for i in range(len(pieces) - 1):
        if pieces[i] >= pieces[i + 1]:
            return False  # Not in increasing order
    return True  # In increasing order

# Function to handle user input and validate it
def get_valid_number_input():
    while True:
        user_input = input("Enter a whole number: ")
        if user_input.isdigit():  # Ensure the input is a whole number
            return user_input
        else:
            print("Invalid input. Please enter a valid whole number.")

def get_valid_piece_length_input(number_length):
    while True:
        user_input = input(f"Enter the number of digits in each piece (must divide {number_length}): ")
        if user_input.isdigit():
            piece_length = int(user_input)
            if number_length % piece_length == 0:
                return piece_length
            else:
                print(f"Invalid input. {piece_length} does not divide {number_length} evenly.")
        else:
            print("Invalid input. Please enter a valid whole number.")

def main():
    # Get valid number input from user
    number = get_valid_number_input()
    number_length = len(number)

    # Get valid piece length input from user
    piece_length = get_valid_piece_length_input(number_length)

    # Check if the pieces are in increasing order
    if check_increasing_order(number, piece_length):
        print("The pieces are in increasing order.")
    else:
        print("The pieces are not in increasing order.")

# Call the main function to run the program
if __name__ == "__main__":
    main()
```

```python
# proj04.py - Factorial, Sine, and Cosine Calculation Program

import math

# Function to calculate the factorial of N
def factorial(n):
    if n < 0:
        return None  # Factorial is not defined for negative numbers
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Function to calculate sine using the power series approximation
def sine(x):
    epsilon = 1.0e-8
    x = x % (2 * math.pi)  # Normalize the angle to [0, 2π)
    result = 0.0
    term = x  # First term of the series
    k = 0

    while abs(term) >= epsilon:
        result += term
        k += 1
        term = (-1)**k * (x**(2*k + 1)) / factorial(2*k + 1)

    return result

# Function to calculate cosine using the power series approximation
def cosine(x):
    epsilon = 1.0e-8
    x = x % (2 * math.pi)  # Normalize the angle to [0, 2π)
    result = 1.0  # First term of the series
    term = 1.0  # First term of the series
    k = 1

    while abs(term) >= epsilon:
        term = (-1)**k * (x**(2*k)) / factorial(2*k)
        result += term
        k += 1

    return result

# Function to display the menu
def display_menu():
    print("\nMenu Options:")
    print("A. Display the value of the factorial of N.")
    print("B. Display the value of the sine of X.")
    print("C. Display the value of the cosine of X.")
    print("M. Display the menu of options.")
    print("Q. Exit from the program.")

def main():
    while True:
        display_menu()

        option = input("\nSelect an option: ").lower()

        if option == 'q':
            print("Exiting the program.")
            break

        elif option == 'a':
            try:
                n = int(input("Enter a non-negative integer N: "))
                if n < 0:
                    print("Factorial is not defined for negative numbers. Try again.")
                    continue
                print(f"The factorial of {n} is {factorial(n)}.")
            except ValueError:
                print("Invalid input. Please enter a valid integer.")

        elif option == 'b':
            try:
                x = float(input("Enter the value of X (in radians): "))
                print(f"The sine of {x} is approximately {sine(x)}.")
            except ValueError:
                print("Invalid input. Please enter a valid number.")

        elif option == 'c':
            try:
                x = float(input("Enter the value of X (in radians): "))
                print(f"The cosine of {x} is approximately {cosine(x)}.")
            except ValueError:
                print("Invalid input. Please enter a valid number.")

        elif option == 'm':
            display_menu()

        else:
            print("Invalid option. Please select a valid option.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
# proj03.py - Additive and Multiplicative Persistence Calculator

# Program Header
# Section: [Your Section Here]
# Date: [Current Date]
# Project Number: 03
# Description: This program calculates the additive and multiplicative persistence of a given integer and their roots.

def additive_persistence(n):
    # Calculate the additive persistence and additive root
    count = 0
    while n >= 10:  # Continue until we get a single digit
        n = sum(int(digit) for digit in str(n))  # Sum the digits
        count += 1
    return count, n

def multiplicative_persistence(n):
    # Calculate the multiplicative persistence and multiplicative root
    count = 0
    while n >= 10:  # Continue until we get a single digit
        product = 1
        for digit in str(n):  # Multiply the digits
            product *= int(digit)
        n = product
        count += 1
    return count, n

def main():
    while True:
        # Gather input from the user
        try:
            number = int(input("Enter an integer (negative number to quit): "))

            if number < 0:
                print("Exiting the program.")
                break

            if number < 10:
                # If the number is a single digit, set persistence and roots to 0 and the number itself
                print(f"Additive persistence: 0, Additive root: {number}")
                print(f"Multiplicative persistence: 0, Multiplicative root: {number}")
                continue

            # Calculate additive persistence and root
            add_persistence, add_root = additive_persistence(number)
            # Calculate multiplicative persistence and root
            mult_persistence, mult_root = multiplicative_persistence(number)

            # Output the results
            print(f"Additive persistence: {add_persistence}, Additive root: {add_root}")
            print(f"Multiplicative persistence: {mult_persistence}, Multiplicative root: {mult_root}")

        except ValueError:
            print("Invalid input. Please enter a valid integer.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
import math

# Function to check if a number is a square
def is_square(num):
    root = math.sqrt(num)
    return root.is_integer()

# Function to generate triangular numbers
def generate_triangular_numbers(limit):
    triangular_numbers = []
    n = 1
    while True:
        t_num = (n * (n + 1)) // 2  # Triangular number formula: T_n = n * (n + 1) / 2
        if t_num > limit:
            break
        triangular_numbers.append(t_num)
        n += 1
    return triangular_numbers

# Main function to solve the problem
def find_triangular_sum(square_num):
    # Check if the input number is a square number
    if not is_square(square_num):
        print(f"The number {square_num} is not a perfect square!")
        return

    # Generate triangular numbers up to the square number
    triangular_numbers = generate_triangular_numbers(square_num)

    # Look for two triangular numbers whose sum is equal to the square number
    for i in range(len(triangular_numbers)):
        for j in range(i, len(triangular_numbers)):
            if triangular_numbers[i] + triangular_numbers[j] == square_num:
                print(f"The two triangular numbers that sum to {square_num} are {triangular_numbers[i]} and {triangular_numbers[j]}")
                return
    print(f"No two triangular numbers sum to {square_num}.")

# User Input and Execution
def main():
    try:
        square_num = int(input("Enter a square number: "))
        find_triangular_sum(square_num)
    except ValueError:
        print("Invalid input. Please enter a valid integer.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
# proj03.py - Russian Peasant Multiplication

# Start of the program with comments
# This program multiplies two integers using the Russian Peasant (Ancient Egyptian) multiplication algorithm.
# It handles positive and negative numbers, and will allow the user to repeat the process.
# The program also indicates whether the final product is positive, negative, or zero.

def russian_peasant_multiplication(a, b):
    # Initialize the product
    product = 0
    steps = []  # To store the intermediate steps for display

    # Process the multiplication using the Russian Peasant algorithm
    while b > 0:
        steps.append((a, b))  # Save the current values of A and B for each step
        if b % 2 == 1:  # Check if B is odd
            product += a
            steps.append(f"B was odd, we add A to make the product: {product}")

        # Double A and halve B (integer division)
        a *= 2
        b //= 2

    # After processing all steps, return the product and the steps
    return product, steps


def main():
    while True:
        try:
            # Get user input
            user_input = input("Please input the 2 numbers separated by a space: ")
            a, b = map(int, user_input.split())
            print(f"A = {a} and B = {b}")

            # Call the Russian Peasant multiplication function
            product, steps = russian_peasant_multiplication(a, b)

            # Print each step
            for step in steps:
                if isinstance(step, tuple):
                    print(f"A = {step[0]} and B = {step[1]}")
                else:
                    print(step)

            # Determine the sign of the product
            if product > 0:
                print("Product is positive")
            elif product < 0:
                print("Product is negative")
            else:
                print("Product is zero")

            # Display the final product
            print(f"The product of the two numbers is: {product}")

            # Ask if the user wants to perform another calculation
            continue_input = input("Do you want to continue?(y/n): ").lower()
            if continue_input == 'y':
                print("=" * 60)
                continue
            elif continue_input == 'n':
                print("Goodbye! Exiting the program.")
                break
            else:
                print("Bad input, quitting.")
                break

        except ValueError:
            print("Invalid input. Please enter two integers separated by a space.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
# proj04.py - Bank Transaction Program

# This program reads customer data from a master file, processes transactions, and updates the file.
# Author: [Your Name]
# Date: [Current Date]
# Course: [Your Course]

# Function to get account number from a record line
def get_number(one_line):
    return one_line.split()[0]

# Function to get account balance from a record line
def get_balance(one_line):
    return float(one_line.split()[1])

# Function to get the account holder's name from a record line
def get_name(one_line):
    return ' '.join(one_line.split()[2:])

# Function to compare two floating-point numbers for equality
def equal_floats(x_flt, y_flt):
    return abs(x_flt - y_flt) < 1.0e-8

# Function to handle the customer transactions
def process_transactions():
    # Prompt for the file name prefix
    prefix = input("Enter the file name prefix: ").strip()

    old_file_name = f"{prefix}.old.txt"
    new_file_name = f"{prefix}.new.txt"

    try:
        # Open the old file for reading and the new file for writing
        old_file = open(old_file_name, "r")
        new_file = open(new_file_name, "w")

    except IOError:
        print(f"Error: Unable to open file(s). Please check file path or permissions.")
        return

    # Read the old master file line by line
    for line in old_file:
        if line.strip() == "999999":
            break  # End of records

        # Extract customer details from the line
        account_number = get_number(line)
        account_balance = get_balance(line)
        account_name = get_name(line)

        # Display customer information
        print(f"Account Number: {account_number}, Name: {account_name}, Balance: {account_balance:.2f}")

        # Process transactions for the customer
        while True:
            transaction = input("Enter transaction (d - deposit, w - withdrawal, c - close, a - advance): ").strip().lower()

            if transaction == "d":  # Deposit
                try:
                    amount = float(input("Enter deposit amount: ").strip())
                    if amount > 0 and account_balance + amount <= 9999999.99:
                        account_balance += amount
                        print(f"Deposited {amount:.2f}. New balance: {account_balance:.2f}")
                    else:
                        print("Error: Invalid deposit amount.")
                except ValueError:
                    print("Error: Invalid amount entered.")

            elif transaction == "w":  # Withdrawal
                try:
                    amount = float(input("Enter withdrawal amount: ").strip())
                    if amount > 0 and account_balance - amount >= 0.00:
                        account_balance -= amount
                        print(f"Withdrawn {amount:.2f}. New balance: {account_balance:.2f}")
                    else:
                        print("Error: Invalid withdrawal amount or insufficient funds.")
                except ValueError:
                    print("Error: Invalid amount entered.")

            elif transaction == "c":  # Close Account
                if equal_floats(account_balance, 0.00):
                    print("Closing account...")
                    break  # Skip this customer and move to next
                else:
                    print("Error: Account balance must be zero to close the account.")

            elif transaction == "a":  # Advance to next customer
                break  # Move to next customer

            else:
                print("Error: Invalid transaction code.")

        # Write the updated record to the new master file if the account is not closed
        if not equal_floats(account_balance, 0.00):
            new_file.write(f"{account_number} {account_balance:.2f} {account_name}\n")

    # Write the sentinel value at the end of the new master file
    new_file.write("999999\n")

    # Close the files
    old_file.close()
    new_file.close()

    print("Transactions completed. Updated master file has been created.")

# Main function to run the program
if __name__ == "__main__":
    process_transactions()
```

```python
# proj04.py - The 196 Algorithm: Check for Palindromes and Lychrel Numbers
# Author: [Your Name]
# Date: [Date]
# Project: Programming Project #4

def is_palindrome(n):
    """Check if a number is a palindrome."""
    str_n = str(n)
    return str_n == str_n[::-1]

def reverse_number(n):
    """Return the reverse of a number."""
    return int(str(n)[::-1])

def check_lychrel(n, max_iterations=60):
    """Check if a number is a Lychrel number."""
    for _ in range(max_iterations):
        n = n + reverse_number(n)
        if is_palindrome(n):
            return False, n  # Not a Lychrel number, returns the palindrome
    return True, n  # Lychrel number

def main():
    # Introduction
    print("This program checks for palindromes and Lychrel numbers in a given range.")

    # Step 1: Prompt the user for two integers (the range)
    start = int(input("Enter the starting integer of the range: "))
    end = int(input("Enter the ending integer of the range: "))

    # Initialize counters
    natural_palindromes = 0
    non_lychrel_numbers = 0
    lychrel_numbers = 0

    # Step 2: Check each number in the given range
    for num in range(start, end + 1):
        if is_palindrome(num):  # If it's already a palindrome
            natural_palindromes += 1
        else:
            is_lychrel, result = check_lychrel(num)
            if is_lychrel:
                lychrel_numbers += 1
                print(f"{num} is a Lychrel number.")
            else:
                non_lychrel_numbers += 1

    # Step 3: Report the statistics
    print("\nStatistics:")
    print(f"Number of natural palindromes: {natural_palindromes}")
    print(f"Number of non-Lychrel numbers: {non_lychrel_numbers}")
    print(f"Number of Lychrel numbers: {lychrel_numbers}")

if __name__ == "__main__":
    main()
```

```python
# proj04.py - Mastermind Guessing Game
# Author: [Your Name]
# Date: [Date]
# Project: Programming Project #4

import random

def is_valid_secret_number(secret_number):
    """Checks if the secret number is a valid 5-digit number with no repeated digits."""
    return len(secret_number) == 5 and secret_number.isdigit() and len(set(secret_number)) == 5

def generate_random_secret_number():
    """Generates a random 5-digit number with no repeated digits."""
    digits = random.sample('0123456789', 5)
    return ''.join(digits)

def give_feedback(secret_number, guess):
    """Gives feedback on how many digits are correct and how many are in the correct position."""
    correct_digits = sum(1 for x, y in zip(secret_number, guess) if x == y)
    correct_positions = sum(1 for x in guess if x in secret_number) - correct_digits
    return correct_positions, correct_digits

def main():
    # Introduction
    print("Welcome to the Mastermind Guessing Game!")

    # Step 1: Prompt for game mode (admin provided or random secret)
    game_mode = input("Do you want the administrator to provide the secret number? (yes/no): ").lower()

    if game_mode == 'yes':
        # Admin provides the secret number
        while True:
            secret_number = input("Enter a 5-digit secret number (no repeated digits): ")
            if is_valid_secret_number(secret_number):
                print("\n" * 50)  # Hide the secret number by printing blank lines
                break
            else:
                print("Invalid input. Please enter a valid 5-digit number with no repeated digits.")
    else:
        # Generate a random secret number
        secret_number = generate_random_secret_number()
        print("\n" * 50)  # Hide the secret number

    # Step 2: Ask the player for guesses and give feedback
    max_tries = 10  # Set a maximum number of tries
    guess_count = 0

    print(f"Game started! You have {max_tries} attempts to guess the secret number.")

    while guess_count < max_tries:
        guess = input(f"Attempt {guess_count + 1}/{max_tries} - Enter your 5-digit guess: ")

        # Validate guess
        if len(guess) != 5 or not guess.isdigit():
            print("Invalid guess. Please enter a 5-digit number.")
            continue

        guess_count += 1

        # Step 3: Check feedback
        correct_positions, correct_digits = give_feedback(secret_number, guess)
        print(f"Feedback: {correct_digits} digits are correct and {correct_positions} digits are in the correct position.")

        # Step 4: Check if the guess is correct
        if guess == secret_number:
            print(f"Congratulations! You've guessed the secret number {secret_number} in {guess_count} attempts.")
            break
    else:
        # If player exceeds max tries
        print(f"Sorry, you've used all {max_tries} attempts. The secret number was {secret_number}. You lost!")

    # Option to quit the game
    quit_game = input("Do you want to quit the game? (yes/no): ").lower()
    if quit_game == 'yes':
        print(f"You lost! The secret number was {secret_number}. You made {guess_count} attempts.")

if __name__ == "__main__":
    main()
```

```python
# proj04.py - Hangman Game
# Author: [Your Name]
# Date: [Date]
# Project: Programming Project #4

import random

# Function to display the hangman game status
def display_game_state(word, guessed_letters, incorrect_guesses, remaining_chances):
    # Display the word with blanks for unguessed letters
    display_word = ''
    for letter in word:
        if letter in guessed_letters or not letter.isalpha():  # Show letter or non-alphabet character (e.g., spaces, dashes)
            display_word += letter + ' '
        else:
            display_word += '_ '

    print(f"Word: {display_word.strip()}")
    print(f"Incorrect guesses: {', '.join(incorrect_guesses)}")
    print(f"Remaining chances: {remaining_chances}")

def main():
    # Introduction and instructions
    print("Welcome to the Hangman game!")
    print("You will have 6 chances to guess the correct letters of the word.")
    print("Each time you guess wrong, you lose one chance.")
    print("Try to guess the word or phrase within your chances. Good luck!\n")

    # Step 1: Ask the user to enter the word/phrase to be guessed
    word = input("Enter the word or phrase to guess: ").lower()

    # Step 2: Setup initial game state
    guessed_letters = set()  # To keep track of correctly guessed letters
    incorrect_guesses = []  # To store the incorrect guesses
    remaining_chances = 6  # Total chances allowed
    guessed_word = ['_'] * len(word)  # Placeholder for the word being guessed

    # Step 3: Game loop
    while remaining_chances > 0:
        display_game_state(word, guessed_letters, incorrect_guesses, remaining_chances)

        # Ask for the player's guess
        guess = input("\nEnter a letter: ").lower()

        # Check for invalid input (non-alphabetical or already guessed letters)
        if not guess.isalpha() or len(guess) != 1:
            print("Invalid input. Please enter a single letter.")
            continue
        if guess in guessed_letters or guess in incorrect_guesses:
            print(f"You've already guessed '{guess}'. Try another letter.")
            continue

        # Step 4: Check if the guess is correct
        if guess in word:
            guessed_letters.add(guess)
            print(f"Good guess! '{guess}' is in the word.")

            # Update the displayed word
            for index, letter in enumerate(word):
                if letter == guess:
                    guessed_word[index] = guess
        else:
            incorrect_guesses.append(guess)
            remaining_chances -= 1
            print(f"Oops! '{guess}' is not in the word.")

        # Step 5: Check if the word has been fully guessed
        if ''.join(guessed_word) == word:
            print(f"Congratulations! You've guessed the word: {word}")
            break

    # If no chances remain, the player loses
    if remaining_chances == 0:
        print(f"Sorry, you've run out of chances. The word was: {word}")

    print("\nGame Over!")

if __name__ == "__main__":
    main()
```

```python
def rotate_alphabet(rotation):
    """Return a rotated version of the alphabet based on the given rotation."""
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    return alphabet[rotation:] + alphabet[:rotation]

def encode(text, rotation):
    """Encode the given text using the rotation cipher."""
    normal_alphabet = 'abcdefghijklmnopqrstuvwxyz'
    rotated_alphabet = rotate_alphabet(rotation)

    encoded_text = ''
    for char in text:
        if char.islower():
            index = normal_alphabet.index(char)
            encoded_text += rotated_alphabet[index]
        else:
            encoded_text += char  # Non-alphabet characters remain unchanged.

    return encoded_text

def decode(encoded_text, target_word):
    """Decode the encoded text by testing all possible rotations from 1 to 25."""
    normal_alphabet = 'abcdefghijklmnopqrstuvwxyz'

    for rotation in range(1, 26):
        rotated_alphabet = rotate_alphabet(rotation)
        decoded_text = ''

        for char in encoded_text:
            if char.islower():
                index = rotated_alphabet.index(char)
                decoded_text += normal_alphabet[index]
            else:
                decoded_text += char  # Non-alphabet characters remain unchanged.

        if target_word in decoded_text:
            return rotation, decoded_text  # Return the correct rotation and decoded text

    return None, None  # Return None if no valid rotation was found.

def main():
    """Main function to run the cipher program."""
    while True:
        # Prompt for the command
        command = input("Enter command (e to encode, d to decode, q to quit): ").strip().lower()

        if command == 'e':  # Encoding
            text = input("Enter text to encode: ")
            while True:
                try:
                    rotation = int(input("Enter a rotation value (1-25): "))
                    if 1 <= rotation <= 25:
                        break
                    else:
                        print("Rotation must be between 1 and 25.")
                except ValueError:
                    print("Invalid input. Please enter a number between 1 and 25.")

            encoded_text = encode(text, rotation)
            print(f"Encoded text: {encoded_text}")

        elif command == 'd':  # Decoding
            encoded_text = input("Enter the encoded text: ")
            target_word = input("Enter a word to search for in the decoded text: ")

            rotation, decoded_text = decode(encoded_text, target_word)
            if rotation is not None:
                print(f"Rotation: {rotation}")
                print(f"Decoded text: {decoded_text}")
            else:
                print("No valid rotation found that decodes the text to include the target word.")

        elif command == 'q':  # Quit
            print("Thank you for using the cipher program. Goodbye!")
            break

        else:
            print("Invalid command. Please enter 'e', 'd', or 'q'.")

# Run the program
if __name__ == '__main__':
    main()
```
